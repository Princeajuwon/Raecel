# How much time do churned users and non-churned users spend on the network during the day, evening, and night on average?


```python
import pandas as pd

# Load the dataset
file_path = 'telecom_churn.csv' 
churn_data = pd.read_csv(file_path)

print(churn_data.columns)
```

```python
required_columns = ['Churn', 'Total day minutes', 'Total eve minutes', 'Total night minutes']
if all(col in churn_data.columns for col in required_columns):
    churn_averages = churn_data.groupby('Churn')[['Total day minutes', 'Total eve minutes', 'Total night minutes']].mean()
    print("Average time spent on the network by churned and non-churned users:")
    print(churn_averages)
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")
```

# What is the average total charge (day, evening, night, international) for customers who churned versus those who did not?


```python
# Required columns for analysis
required_columns = ['Churn', 'Total day charge', 'Total eve charge', 'Total night charge', 'Total intl charge']

# Check if all required columns are present
if all(col in churn_data.columns for col in required_columns):
    # Group by 'Churn' and calculate the average for the total charges
    churn_avg_charges = churn_data.groupby('Churn')[['Total day charge', 'Total eve charge', 'Total night charge', 'Total intl charge']].mean()

    print("Average total charges (day, evening, night, international) by churn status:")
    print(churn_avg_charges)
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")
```

# How do daytime, evening, and nighttime usage metrics compare among customers who churned vs. those who stayed?


```python
# Required columns for analysis
required_columns = ['Churn', 'Total day minutes', 'Total eve minutes', 'Total night minutes']

# Check if all required columns are present
if all(col in churn_data.columns for col in required_columns):
    # Group by 'Churn' and calculate the average for each usage metric
    churn_avg_usage = churn_data.groupby('Churn')[['Total day minutes', 'Total eve minutes', 'Total night minutes']].mean()

    print("Comparison of daytime, evening, and nighttime usage metrics by churn status:")
    print(churn_avg_usage)
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")
```

# Are there significant differences in churn rates across different states or area codes?


```python


# Required columns for analysis
required_columns = ['Churn', 'State', 'Area code']

# Check if all required columns are present
if all(col in churn_data.columns for col in required_columns):
    # Calculate churn rates by State
  
    state_churn_rates = churn_data.groupby('State')['Churn'].mean() * 100  # Percentage of churned customers
    print("Churn rates by State:")
    print(state_churn_rates.sort_values(ascending=False))

    print("\n")

    # Calculate churn rates by Area Code
    area_code_churn_rates = churn_data.groupby('Area code')['Churn'].mean() * 100  # Percentage o
    print("area_code_churn_rates:")
    print(area_code_churn_rates.sort_values(ascending=False))
    print("\n")
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")
```

# Do customers with an international plan have different usage patterns compared to those without?


```python
# Check if all required columns are present
if all(col in churn_data.columns for col in required_columns):
    # Group by 'Intl Plan' and calculate the average for each usage metric
    intl_plan_usage = churn_data.groupby('International plan')[['Total day minutes', 'Total eve minutes', 'Total night minutes', 'Total intl minutes']].mean()
    print("Usage patterns for customers with and without an international plan:")
    print(intl_plan_usage)
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")  
```

# Who engages customer care the most among the customers?


```python
# Required columns for analysis
required_columns = ['Customer service calls', 'Churn', 'International plan']

# Check if all required columns are present
if all(col in churn_data.columns for col in required_columns):
    # Total customer service calls grouped by churn status
    churn_service_calls = churn_data.groupby('Churn')['Customer service calls'].mean()

    print("Average customer service calls by churn status:")
    print(churn_service_calls)
    print("\n")

    # Total customer service calls grouped by international plan
    intl_service_calls = churn_data.groupby('International plan')['Customer service calls'].mean()

    print("Average customer service calls by international plan status:")
    print(intl_service_calls)
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")
```

# What are the characteristics of loyal customers?


```python
import pandas as pd
file_path = 'telecom_churn.csv'  
churn_data = pd.read_csv(file_path)

# Filter loyal customers (Churn == 0)
loyal_customers = churn_data[churn_data['Churn'] == 0]

# Analyze characteristics
if 'Voice mail plan' in churn_data.columns:
    # Proportion with voicemail plan
    voicemail_plan_loyal = loyal_customers['Voice mail plan'].value_counts(normalize=True) * 100
    print("Voice mail plan distribution among loyal customers:")
    print(voicemail_plan_loyal)
    print("\n")

if 'International plan' in churn_data.columns:
    # Proportion with international plan
    intl_plan_loyal = loyal_customers['International plan'].value_counts(normalize=True) * 100
    print("International plan distribution among loyal customers:")
    print(intl_plan_loyal)
    print("\n")

# Average usage metrics for loyal customers
usage_columns = ['Total day minute', 'Total eve minutes', 'Total night minutes', 'Total intl minutes']
if all(col in churn_data.columns for col in usage_columns):
    avg_usage_loyal = loyal_customers[usage_columns].mean()
    print("Average usage metrics for loyal customers:")
    print(avg_usage_loyal)
    print("\n")

# Average customer service calls for loyal customers
if 'Customer service calls' in churn_data.columns:
    avg_cust_service_calls_loyal = loyal_customers['Customer service calls'].mean()
    print("Average customer service calls for loyal customers:")
    print(avg_cust_service_calls_loyal)
    print("\n")
```
